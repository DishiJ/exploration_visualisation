**a. Fit a linear regression using Python to the above aggregated data (i.e., total number of births registered in Australia over time) and plot the linear fit.**

```python
#A1-2a
import pandas as pd
lp = pd.DataFrame(birth_sum_yearwise)
lp.reset_index(inplace = True)
lp.rename(columns = {0 : 'Birth Sum'} , inplace = True)
```

```python
X = lp.iloc[:, 0:1].values 
y = lp.iloc[:, 1].values 
```

```python
# Fitting Linear Regression to the dataset 
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression 
linn = LinearRegression() 
  
linn.fit(X, y)
y_pred = linn.predict(X)
rmse = np.sqrt(mean_squared_error(y,y_pred))
print(rmse)
r2 = r2_score(y,y_pred)
print(r2)
```

```python
# Visualising the Linear Regression results 
plt.figure(figsize=(12, 8))    

plt.scatter(X, y, color = 'blue') 
  
plt.plot(X, linn.predict(X), color = 'red') 

plt.xlabel("Year",labelpad=15,color='black',fontsize='xx-large')
plt.ylabel("Number of Births",labelpad=15,color='black',fontsize='xx-large')
plt.title("Total number of births registered in Australia over time using Linear Regression",fontsize='xx-large',fontweight =600,pad=20)

  
plt.show() 
```

```python
# Fitting Polynomial Regression to the dataset 
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures 
  
poly = PolynomialFeatures(degree = 4) 
X_poly = poly.fit_transform(X) 
  
poly.fit(X_poly, y) 
lin2 = LinearRegression() 
lin2.fit(X_poly, y)
y_poly_pred = lin2.predict(X_poly)
rmse = np.sqrt(mean_squared_error(y,y_poly_pred))
print(rmse)
r2 = r2_score(y,y_poly_pred)
print(r2)
```

```python
# Visualising the Polynomial Regression results 
plt.figure(figsize=(12, 8))    

plt.scatter(X, y, color = 'blue') 
  
plt.plot(X, lin2.predict(poly.fit_transform(X)), color = 'red') 

plt.xlabel("Year",labelpad=15,color='black',fontsize='xx-large')
plt.ylabel("Total number of Births",labelpad=15,color='black',fontsize='xx-large')
plt.title("Total number of births registered in Australia over time using Polynomial Regression",fontsize='xx-large',fontweight =600,pad=20)
  
plt.show() 
```

```python
deaths=pd.read_csv("Deaths.csv")
births=pd.read_csv("Births.csv")
tfr=pd.read_csv("TFR.csv")
nim=pd.read_csv("NIM.csv")
nom=pd.read_csv("NOM.csv")
```

```python
melt_births = pd.melt(births, id_vars=['Year'],value_vars=['NSW','ACT','QLD','SA','WA','TAS','VIC','NT'],var_name='States',value_name = 'BIRTHS')
melt_deaths = pd.melt(deaths, id_vars=['Year'],value_vars=['NSW','ACT','QLD','SA','WA','TAS','VIC','NT'],var_name='States',value_name = 'DEATHS')
melt_tfr = pd.melt(tfr, id_vars=['Year'],value_vars=['NSW','ACT','QLD','SA','WA','TAS','VIC','NT'],var_name='States',value_name = 'TFR')
melt_nim = pd.melt(nim, id_vars=['Year'],value_vars=['NSW','ACT','QLD','SA','WA','TAS','VIC','NT'],var_name='States',value_name = 'NIM')
melt_nom = pd.melt(nom, id_vars=['Year'],value_vars=['NSW','ACT','QLD','SA','WA','TAS','VIC','NT'],var_name='States',value_name = 'NOM')
```

```python
melt_nom['NIM'] = melt_nim['NIM']
melt_nom['DEATHS'] = melt_deaths['DEATHS']
melt_nom['BIRTHS'] = melt_births['BIRTHS']
melt_nom['TFR'] = melt_tfr['TFR']
final = melt_nom
```
